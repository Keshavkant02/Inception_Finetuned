```python
# ================================================================================================
#                A BEGINNER'S GUIDE TO FINE-TUNING INCEPTION-V3 FOR MEDICAL IMAGING
# ================================================================================================
# This notebook is a complete, end-to-end tutorial on fine-tuning a powerful, pre-trained
# model (Inception-V3) for the task of classifying pneumonia from chest X-ray images.
#
# We will walk through every step, from setting up the environment to evaluating our final model.
# The comments are designed to be highly educational, explaining the "why" behind each code decision.
# ================================================================================================

# --- Core Libraries ---
# These are the essential tools for deep learning, data manipulation, and visualization.
import subprocess
import sys
import gc
import os
import datetime
import warnings

# --- Tidy Up ---
# We'll ignore common warnings from libraries to keep our output clean.
warnings.filterwarnings('ignore')

# ================================================================================================
# PHASE 0: COLAB UTILITIES - KEEPING OUR WORKSHOP RUNNING
# ================================================================================================
# This section contains helper functions to make our Colab experience smooth.

def setup_keep_alive():
    """
    KEEPS OUR COLAB SESSION ALIVE.
    Colab notebooks will automatically disconnect if they are idle for too long.
    This function runs a small piece of Javascript in the background that "clicks" the
    connect button every 60 seconds, tricking Colab into thinking we're still active.
    This is essential for long, unattended training sessions (like running overnight).
    """
    try:
        # Import the display module from IPython to run Javascript
        import IPython.display as display
        # The Javascript code to be executed by the browser
        js_code = '''
            function ClickConnect(){
                console.log("Working - Keeping Colab alive");
                // Find the connect button on the Colab toolbar
                var connectButton = document.querySelector("colab-toolbar-button#connect");
                // If the button exists, click it
                if (connectButton) { connectButton.click(); }
            }
            // Run the ClickConnect function every 60,000 milliseconds (60 seconds)
            setInterval(ClickConnect, 60000);
        '''
        # Display the Javascript to run it
        display.display(display.Javascript(js_code))
        print("üåô Keep-alive activated! Your session is now sleep-friendly.")
    except Exception as e:
        # If something goes wrong (e.g., not running in Colab), print an error.
        print(f"Keep-alive setup failed: {e}")

# Run our keep-alive setup function immediately.
setup_keep_alive()


def install_package(package):
    """
    INSTALLS A PYTHON PACKAGE ONLY IF IT'S NOT ALREADY INSTALLED.
    This is a clean way to manage dependencies without re-installing things every time.
    """
    try:
        # Try to import the package to see if it exists.
        __import__(package.split('==')[0])
        print(f"‚úÖ {package} is already installed. Ready to go!")
    except ImportError:
        # If the import fails, it means the package is not installed.
        print(f"üì¶ Package '{package}' not found. Installing now...")
        # Use pip to install the package quietly (-q).
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package])
        print(f"‚úÖ {package} installed successfully.")

# Install the specific packages needed for this notebook.
install_package("medmnist") # For the PneumoniaMNIST dataset
install_package("seaborn") # For beautiful visualizations
install_package("tqdm") # For progress bars

# --- Import all our necessary libraries for the project ---
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.utils.class_weight import compute_class_weight
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import torchvision.models as models
from tqdm import tqdm
import medmnist
from medmnist import INFO, PneumoniaMNIST

# ================================================================================================
# PHASE 1: SETUP & REPRODUCIBILITY - BUILDING A SOLID FOUNDATION
# ================================================================================================
# Good science is reproducible science. We set "seeds" to ensure that anyone who runs
# this code gets the exact same results. This removes randomness from weight initialization
# and data shuffling, which is crucial for fair experiments.

# Set the master seed for PyTorch (for CPU and GPU) and NumPy.
torch.manual_seed(42)
np.random.seed(42)

# Set the device. This line is magic: it automatically detects if a GPU ('cuda') is
# available and uses it. If not, it falls back to the CPU.
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# If we are using a GPU, apply additional settings for full reproducibility.
if torch.cuda.is_available():
    torch.cuda.manual_seed(42)
    # These two lines ensure that CUDA operations are deterministic.
    # It can slightly slow down training, but it's essential for reproducible results.
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

print(f"üß† Using device: {device.type.upper()}")

# Clean up memory before we start to ensure we have a fresh state.
gc.collect() # Ask the Python garbage collector to free up memory.
if torch.cuda.is_available():
    torch.cuda.empty_cache() # Clear any cached memory on the GPU.

# ================================================================================================
# PHASE 2: DATA PREPARATION - THE MOST IMPORTANT STEP
# ================================================================================================
print("\nüî¨ Preparing the Data Pipeline...")

# --- CRITICAL DETAIL: Use the Correct Input Size ---
# The Inception-V3 model was pre-trained on images of size 299x299.
# We MUST use this same size to get good performance. Using the wrong size is a
# very common and hard-to-debug mistake in transfer learning.
INPUT_SIZE = 299

# --- The Image "Factory": Defining Our Transformations ---
# We define two separate "recipes" (pipelines) for our images. One for training,
# and one for validation/testing. This is a critical distinction.

# The "Creative Artist" recipe for the TRAINING set.
# This applies Data Augmentation to prevent overfitting.
data_transform_train = transforms.Compose([
    transforms.Resize((INPUT_SIZE, INPUT_SIZE)), # Resize to Inception's required input size.
    transforms.RandomRotation(15),             # Randomly rotate the image.
    transforms.RandomHorizontalFlip(),         # Randomly flip the image horizontally.
    transforms.ColorJitter(brightness=0.1, contrast=0.1), # Slightly change brightness/contrast.
    transforms.Grayscale(num_output_channels=3), # The dataset is grayscale, but Inception-V3 expects a 3-channel (RGB) image. This duplicates the grayscale channel 3 times.
    transforms.ToTensor(),                     # Convert the image to a PyTorch Tensor (scales pixels to [0, 1]).
    # Normalize the tensor. These specific mean and std values are from the ImageNet
    # dataset, which is what Inception-V3 was trained on. We MUST use them.
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# The "Strict Scientist" recipe for the VALIDATION and TEST sets.
# IMPORTANT: We do NOT perform data augmentation here. We want an honest
# evaluation of the model's performance on unaltered images.
data_transform_val_test = transforms.Compose([
    transforms.Resize((INPUT_SIZE, INPUT_SIZE)),
    transforms.Grayscale(num_output_channels=3),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# --- Loading the Dataset ---
# The MedMNIST library makes this incredibly easy. We just specify which split
# ('train', 'val', 'test') we want, apply our transform recipe, and download it.
train_dataset = PneumoniaMNIST(split='train', transform=data_transform_train, download=True)
val_dataset = PneumoniaMNIST(split='val', transform=data_transform_val_test, download=True)
test_dataset = PneumoniaMNIST(split='test', transform=data_transform_val_test, download=True)

# --- Creating DataLoaders ---
# The DataLoader is a highly efficient conveyor belt that feeds batches of data to our model.
BATCH_SIZE = 16  # We keep this small to avoid running out of RAM in Colab.
NUM_WORKERS = 2  # Number of parallel CPU processes to load data. Speeds things up.

# The `pin_memory=True` argument can speed up data transfer from CPU to GPU.
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

print(f"\n‚úÖ Data pipeline ready.")
print(f"   Image size: {INPUT_SIZE}x{INPUT_SIZE}")
print(f"   Training batches: {len(train_loader)}, Validation batches: {len(val_loader)}, Test batches: {len(test_loader)}")


# ================================================================================================
# PHASE 3: MODEL DEFINITION - PERFORMING "BRAIN SURGERY"
# ================================================================================================
print("\nüõ†Ô∏è Defining the Model Architecture...")

# We define our model as a Python class for clarity and organization.
class PneumoniaInceptionV3(nn.Module):
    def __init__(self, num_classes=2, pretrained=True):
        super(PneumoniaInceptionV3, self).__init__()

        # --- STEP 1: Load the pre-trained Inception-V3 model ---
        # `pretrained=True` downloads the weights learned from the ImageNet dataset.
        # `aux_logits=True` is REQUIRED to match the structure of the pre-trained weights.
        # The base Inception-V3 model has an "auxiliary" head that was used during its original
        # training. We must load it, even if we won't use it.
        self.inception = models.inception_v3(pretrained=pretrained, aux_logits=True)

        # --- STEP 2: Deactivate the Auxiliary Head ---
        # Now that the model is loaded correctly, we can effectively "turn off" the
        # auxiliary head by replacing it with an `nn.Identity` layer. This layer
        # just passes the input through without doing anything.
        self.inception.AuxLogits = nn.Identity()

        # --- STEP 3: Freeze the Base Model ---
        # This is the core of Transfer Learning. We do NOT want to re-train the entire
        # model from scratch. We freeze the pre-trained layers to keep their powerful
        # feature-extracting knowledge.
        for param in self.inception.parameters():
            param.requires_grad = False

        # --- STEP 4: Prepare for Surgery - Get Input Features ---
        # We need to know the size of the input to the final layer so we can attach
        # our own custom layer correctly.
        num_ftrs = self.inception.fc.in_features

        # --- STEP 5: Attach Our New "Head" ---
        # We replace the original final layer (`self.inception.fc`) with our own `nn.Sequential`
        # block. ONLY this new block will be trainable (`requires_grad` is True by default for new layers).
        self.inception.fc = nn.Sequential(
            nn.Linear(num_ftrs, 512),       # A new fully-connected layer to learn high-level patterns.
            nn.ReLU(inplace=True),          # A standard activation function.
            nn.Dropout(0.5),                # Dropout is a regularization technique to prevent overfitting.
            nn.Linear(512, num_classes)     # The final output layer. Must have `num_classes` neurons.
        )

    def forward(self, x):
        """Defines the forward pass of the model."""
        # When Inception-V3 is in training mode with aux_logits, it returns two outputs:
        # the main output and the auxiliary output.
        x = self.inception(x)

        # This `if` statement handles the original model's tuple output.
        # Even though we've replaced the AuxLogits, this is good practice for robustness.
        # It ensures that if we are training, we only take the first (main) output.
        if self.training and isinstance(x, tuple):
             x = x[0]
        return x

# --- Create an instance of our model ---
model = PneumoniaInceptionV3(num_classes=2, pretrained=True)
# Move the model to the correct device (GPU or CPU).
model = model.to(device)

# --- Verify Our Setup ---
# Let's count the parameters to confirm our freezing strategy worked.
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"‚úÖ Model created successfully.")
print(f"   Total parameters: {total_params:,}")
print(f"   Trainable parameters: {trainable_params:,} (SUCCESS! Only our new head is trainable)")


# ================================================================================================
# PHASE 4: TRAINING CONFIGURATION - SETTING THE RULES OF LEARNING
# ================================================================================================
print("\n‚öôÔ∏è Configuring the Training Process...")

# --- STRATEGY 1: Handling Imbalanced Data with Class Weights ---
# Our dataset might have more "normal" images than "pneumonia" images. If we don't
# account for this, the model might just learn to always predict the majority class.
# `compute_class_weight` calculates weights that give more importance to the
# minority class during loss calculation.
train_labels = train_dataset.labels.flatten()
class_weights = compute_class_weight('balanced', classes=np.unique(train_labels), y=train_labels)
class_weights = torch.tensor(class_weights, dtype=torch.float32).to(device)
print(f"Computed Class Weights: {class_weights.cpu().numpy()}. The minority class will be given more importance.")

# --- The Loss Function (The Judge) ---
# We use CrossEntropyLoss, the standard for classification. We pass our computed
# class weights to it.
criterion = nn.CrossEntropyLoss(weight=class_weights)

# --- The Optimizer (The Teacher) ---
# Adam is a modern, robust optimizer. We use `filter` to ensure we are only
# passing the *trainable* parameters (our new head) to the optimizer.
# `weight_decay` is another form of regularization (L2 penalty).
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-4, weight_decay=1e-5)

# --- The Learning Rate Scheduler (The Pace Controller) ---
# This scheduler will automatically reduce the learning rate if the validation loss
# stops improving. It's like a coach telling the team to slow down and be more
# careful when they hit a plateau.
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=2, factor=0.2, verbose=True)


# ================================================================================================
# PHASE 5: THE TRAINING & VALIDATION LOOPS - THE MAIN EVENT
# ================================================================================================
# We define two functions: one for a full training epoch and one for a full validation epoch.
# This keeps the main training loop clean and easy to read.

def train_epoch(model, loader, criterion, optimizer, device):
    """Runs one full epoch of training."""
    model.train()  # Set the model to training mode (enables Dropout, etc.)
    running_loss = 0.0
    correct = 0
    total = 0
    # Use tqdm for a nice progress bar
    pbar = tqdm(loader, desc='Training')
    for images, labels in pbar:
        # Move data to the correct device
        images, labels = images.to(device), labels.squeeze().long().to(device)

        # Standard training steps:
        optimizer.zero_grad()      # 1. Clear old gradients
        outputs = model(images)    # 2. Get model predictions (Forward Pass)
        loss = criterion(outputs, labels) # 3. Calculate the loss
        loss.backward()            # 4. Calculate gradients (Backward Pass)
        optimizer.step()           # 5. Update the weights

        # Update statistics for logging
        running_loss += loss.item() * images.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        # Update the progress bar with live stats
        pbar.set_postfix({'Loss': f'{running_loss/total:.4f}', 'Acc': f'{100.*correct/total:.2f}%'})

    return running_loss / total, 100. * correct / total

def validate(model, loader, criterion, device):
    """Runs one full epoch of validation."""
    model.eval()   # Set the model to evaluation mode (disables Dropout, etc.)
    val_loss = 0.0
    correct = 0
    total = 0
    # We don't need to calculate gradients during validation, which saves memory and computation.
    with torch.no_grad():
        pbar = tqdm(loader, desc='Validation')
        for images, labels in pbar:
            images, labels = images.to(device), labels.squeeze().long().to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)

            val_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            pbar.set_postfix({'Loss': f'{val_loss/total:.4f}', 'Acc': f'{100.*correct/total:.2f}%'})

    return val_loss / total, 100. * correct / total

# --- The Main Training Loop ---
EPOCHS = 10 # Number of full passes through the training data
best_val_loss = float('inf') # Initialize best loss to infinity
history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []} # To store our training history
log_file = f"training_log_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.txt" # A log file to save progress

def log_progress(message):
    """A helper function to print messages and also save them to a file."""
    timestamp = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    log_msg = f"[{timestamp}] {message}"
    print(log_msg)
    with open(log_file, 'a') as f: f.write(log_msg + '\n')

log_progress("üöÄ Starting unattended training session...")
for epoch in range(EPOCHS):
    log_progress(f"--- Epoch {epoch+1}/{EPOCHS} ---")

    # Run one epoch of training and validation
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = validate(model, val_loader, criterion, device)
    # The scheduler checks the validation loss to decide if it should lower the learning rate.
    scheduler.step(val_loss)

    # Store the results for this epoch
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)

    # --- Save the Best Model ---
    # This is a crucial step. We only save the model if its performance on the
    # validation set has improved. This ensures we always have the best version.
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), 'best_model.pth')
        log_progress(f"üéâ NEW BEST MODEL SAVED! Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")

    log_progress(f"Epoch {epoch+1} Summary: Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}% | Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")
    # Clean up memory at the end of each epoch
    gc.collect()
    if torch.cuda.is_available(): torch.cuda.empty_cache()

log_progress("üéâ TRAINING COMPLETED SUCCESSFULLY!")


# ================================================================================================
# PHASE 6: FINAL EVALUATION & VISUALIZATION - THE REPORT CARD
# ================================================================================================
print("\nüìä Evaluating final model on the sacred Test Set...")
# Load the weights of the best model we saved during training.
model.load_state_dict(torch.load('best_model.pth'))

def evaluate_model(model, loader, device):
    """Evaluates the model on a given dataset and returns all predictions and labels."""
    model.eval() # Set to evaluation mode
    all_preds, all_probs, all_labels = [], [], []
    with torch.no_grad():
        for images, labels in tqdm(loader, desc='Testing'):
            images, labels = images.to(device), labels.to(device)
            # Get model outputs (logits)
            outputs = model(images)
            # Convert logits to probabilities using softmax
            probabilities = torch.softmax(outputs, dim=1)
            # Get the final prediction (the class with the highest probability)
            _, predicted = torch.max(outputs, 1)
            # Append results to our lists
            all_preds.extend(predicted.cpu().numpy())
            all_probs.extend(probabilities.cpu().numpy())
            all_labels.extend(labels.squeeze().cpu().numpy())

    return np.array(all_preds), np.array(all_probs), np.array(all_labels)

# Run evaluation on our test loader
test_preds, test_probs, test_labels = evaluate_model(model, test_loader, device)

# --- Calculate Final Performance Metrics ---
# These are the key numbers that you would put in a report or presentation.
accuracy = np.mean(test_preds == test_labels)
report = classification_report(test_labels, test_preds, target_names=['Normal', 'Pneumonia'])
cm = confusion_matrix(test_labels, test_preds)
# For binary classification, ROC AUC is a very important metric. It measures the
# model's ability to distinguish between the two classes.
roc_auc = roc_auc_score(test_labels, test_probs[:, 1])

# --- Display Final Report ---
results_text = f"""
üéØ FINAL TEST RESULTS:
‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê
- Accuracy: {accuracy:.4f}
- ROC AUC Score: {roc_auc:.4f}

{report}
"""
print(results_text)
log_progress(results_text) # Also save the final results to our log file.

# ================================================================================================
# PHASE 7: VISUALIZATION - TELLING THE STORY WITH DATA
# ================================================================================================
# A good visualization is worth a thousand numbers.

# Create a large figure to hold all our plots
plt.figure(figsize=(20, 5))

# Plot 1: Loss vs. Epochs
plt.subplot(1, 4, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Val Loss')
plt.title('Loss vs. Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plot 2: Accuracy vs. Epochs
plt.subplot(1, 4, 2)
plt.plot(history['train_acc'], label='Train Accuracy')
plt.plot(history['val_acc'], label='Val Accuracy')
plt.title('Accuracy vs. Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()

# Plot 3: Confusion Matrix
# This tells us where the model is getting confused.
plt.subplot(1, 4, 3)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal', 'Pneumonia'], yticklabels=['Normal', 'Pneumonia'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')

# Plot 4: ROC Curve
# This visualizes the trade-off between true positive rate and false positive rate.
plt.subplot(1, 4, 4)
fpr, tpr, _ = roc_curve(test_labels, test_probs[:, 1])
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.3f}')
plt.plot([0, 1], [0, 1], 'k--') # A line representing a random guess
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()

# Adjust layout to prevent plots from overlapping
plt.tight_layout()
# Show the plots
plt.show()

log_progress("üéâ ALL PROCESSES COMPLETED SUCCESSFULLY!")
```

```python
import subprocess
import sys
import gc
import os
import datetime
import warnings
warnings.filterwarnings('ignore')

# ================================================================================================
#                      FIXED & IMPROVED "SLEEP-FRIENDLY" SCRIPT
# ================================================================================================
# Self-Critique Summary:
# - Original code had a complex, error-prone custom Dataset class. This is now removed.
# - Original code used the wrong input size (224x224). Corrected to 299x299 for Inception-V3.
# - The entire data preprocessing pipeline is now a single, standard `transforms.Compose` block.
# - Model definition is simplified for clarity and stability.
# ================================================================================================

def setup_keep_alive():
    """Keep Colab alive during long training sessions"""
    try:
        import IPython.display as display
        display.display(display.Javascript('''
            function ClickConnect(){
                console.log("Working - Keeping Colab alive");
                var connectButton = document.querySelector("colab-toolbar-button#connect");
                if (connectButton) { connectButton.click(); }
            }
            setInterval(ClickConnect, 60000);
        '''))
        print("üåô Keep-alive activated! Colab will stay connected during training.")
    except Exception as e:
        print(f"Keep-alive setup failed: {e}")

setup_keep_alive()

# Install required packages
def install_package(package):
    """Install package if not already installed"""
    try:
        __import__(package.split('==')[0])
        print(f"‚úÖ {package} already installed")
    except ImportError:
        print(f"üì¶ Installing {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package])
        print(f"‚úÖ {package} installed successfully")

install_package("medmnist")
install_package("seaborn")
install_package("tqdm")

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.utils.class_weight import compute_class_weight
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import torchvision.models as models
from tqdm import tqdm
import medmnist
from medmnist import INFO, PneumoniaMNIST

# Set seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
if torch.cuda.is_available():
    torch.cuda.manual_seed(42)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

print(f"Using device: {device}")
gc.collect()
if torch.cuda.is_available(): torch.cuda.empty_cache()

# ================================================================================================
# PHASE 1: DATA PREPARATION (SIMPLIFIED & CORRECTED)
# ================================================================================================
print("üî¨ Preparing Data Pipeline...")

# --- CRITICAL FIX: Use the correct input size for Inception-V3 (299x299) ---
INPUT_SIZE = 299

# --- CRITICAL FIX: A single, standard transform pipeline ---
# This pipeline does everything correctly:
# 1. Resizes to the required 299x299.
# 2. Performs data augmentation (for training set only).
# 3. Converts the single-channel grayscale image to a 3-channel image.
# 4. Converts the image to a PyTorch Tensor (and scales pixel values to [0, 1]).
# 5. Normalizes the tensor with ImageNet stats, as expected by the pre-trained model.

data_transform_train = transforms.Compose([
    transforms.Resize((INPUT_SIZE, INPUT_SIZE)),
    transforms.RandomRotation(15),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.1, contrast=0.1),
    transforms.Grayscale(num_output_channels=3), # Ensure 3 channels
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

data_transform_val_test = transforms.Compose([
    transforms.Resize((INPUT_SIZE, INPUT_SIZE)),
    transforms.Grayscale(num_output_channels=3), # Ensure 3 channels
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# --- CRITICAL FIX: Load the dataset and apply the transforms directly. No custom Dataset class needed. ---
train_dataset = PneumoniaMNIST(split='train', transform=data_transform_train, download=True)
val_dataset = PneumoniaMNIST(split='val', transform=data_transform_val_test, download=True)
test_dataset = PneumoniaMNIST(split='test', transform=data_transform_val_test, download=True)

# Data loaders with memory optimization
BATCH_SIZE = 16 # Keep small for Colab free tier
NUM_WORKERS = 2

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

print(f"‚úÖ Data pipeline ready. Image size: {INPUT_SIZE}x{INPUT_SIZE}")
print(f"   Training batches: {len(train_loader)}, Validation batches: {len(val_loader)}, Test batches: {len(test_loader)}")


# PHASE 2: MODEL DEFINITION (CORRECTED FOR aux_logits)
# ================================================================================================
print("üõ†Ô∏è Defining Model Architecture...")

# --- CRITICAL FIX: Respect the `aux_logits` rule for pre-trained models ---
class PneumoniaInceptionV3(nn.Module):
    def __init__(self, num_classes=2, pretrained=True):
        super(PneumoniaInceptionV3, self).__init__()

        # Step 1: Load the model with aux_logits=True. This is REQUIRED by the pre-trained weights.
        self.inception = models.inception_v3(pretrained=pretrained, aux_logits=True)

        # Step 2: Now that the model is loaded, we can disable the auxiliary head
        # by replacing it with an Identity layer that does nothing.
        # This is because we will not use it for training or inference.
        self.inception.AuxLogits = nn.Identity()

        # Step 3: Freeze all layers of the base model
        for param in self.inception.parameters():
            param.requires_grad = False

        # Step 4: Get the number of input features for the final layer
        num_ftrs = self.inception.fc.in_features

        # Step 5: Replace the final fully connected layer (the "head").
        # This new head WILL have its gradients calculated and will be trained.
        self.inception.fc = nn.Sequential(
            nn.Linear(num_ftrs, 512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        # During training, inception_v3 with aux_logits=True returns a tuple (main_output, aux_output).
        # We only care about the main output.
        x = self.inception(x)
        # The line below is not strictly necessary anymore because we replaced AuxLogits,
        # but it's good practice for robustness in case of different model states.
        if self.training and isinstance(x, tuple):
             x = x[0]
        return x

# Create model
model = PneumoniaInceptionV3(num_classes=2, pretrained=True)
model = model.to(device)

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"‚úÖ Model created successfully.")
print(f"   Total parameters: {total_params:,}")
print(f"   Trainable parameters: {trainable_params:,} (Only the final classifier head)")

# ================================================================================================
# PHASE 3: TRAINING SETUP (Unchanged Logic, Adapted to New Data Loading)
# ================================================================================================
print("‚öôÔ∏è Configuring Training...")

# Calculate class weights to handle imbalance
# This is a valid strategy and is kept from the original code.
train_labels = train_dataset.labels.flatten()
class_weights = compute_class_weight('balanced', classes=np.unique(train_labels), y=train_labels)
class_weights = torch.tensor(class_weights, dtype=torch.float32).to(device)
print(f"Class weights computed for loss function: {class_weights.cpu().numpy()}")

# Loss function and optimizer
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=1e-4, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=2, factor=0.2, verbose=True)

# ================================================================================================
# PHASE 4: TRAINING & VALIDATION LOOPS (Largely Unchanged)
# ================================================================================================
# These functions are well-structured and can be kept.

def train_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    pbar = tqdm(loader, desc='Training')
    for images, labels in pbar:
        images, labels = images.to(device), labels.squeeze().long().to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        pbar.set_postfix({'Loss': f'{running_loss/total:.4f}', 'Acc': f'{100.*correct/total:.2f}%'})

    return running_loss / total, 100. * correct / total

def validate(model, loader, criterion, device):
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        pbar = tqdm(loader, desc='Validation')
        for images, labels in pbar:
            images, labels = images.to(device), labels.squeeze().long().to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)

            val_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            pbar.set_postfix({'Loss': f'{val_loss/total:.4f}', 'Acc': f'{100.*correct/total:.2f}%'})

    return val_loss / total, 100. * correct / total

# Training loop with enhanced logging
EPOCHS = 10
best_val_loss = float('inf')
history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
log_file = f"training_log_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"

def log_progress(message):
    timestamp = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    log_msg = f"[{timestamp}] {message}"
    print(log_msg)
    with open(log_file, 'a') as f: f.write(log_msg + '\n')

log_progress("üöÄ Starting unattended training session...")
for epoch in range(EPOCHS):
    log_progress(f"--- Epoch {epoch+1}/{EPOCHS} ---")

    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = validate(model, val_loader, criterion, device)
    scheduler.step(val_loss)

    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), 'best_model.pth')
        log_progress(f"‚úÖ NEW BEST MODEL SAVED! Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")

    log_progress(f"Epoch {epoch+1} Summary: Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}% | Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")
    gc.collect()
    if torch.cuda.is_available(): torch.cuda.empty_cache()

log_progress("üéâ TRAINING COMPLETED SUCCESSFULLY!")

# ================================================================================================
# PHASE 5: EVALUATION & VISUALIZATION (Unchanged Logic)
# ================================================================================================
print("üìä Evaluating final model on test set...")
model.load_state_dict(torch.load('best_model.pth'))

def evaluate_model(model, loader, device):
    model.eval()
    all_preds, all_probs, all_labels = [], [], []
    with torch.no_grad():
        for images, labels in tqdm(loader, desc='Testing'):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)

            probabilities = torch.softmax(outputs, dim=1)
            _, predicted = torch.max(outputs, 1)

            all_preds.extend(predicted.cpu().numpy())
            all_probs.extend(probabilities.cpu().numpy())
            all_labels.extend(labels.squeeze().cpu().numpy())

    return np.array(all_preds), np.array(all_probs), np.array(all_labels)

test_preds, test_probs, test_labels = evaluate_model(model, test_loader, device)

# Calculate metrics (These are your key results for the report)
accuracy = np.mean(test_preds == test_labels)
report = classification_report(test_labels, test_preds, target_names=['Normal', 'Pneumonia'])
cm = confusion_matrix(test_labels, test_preds)
roc_auc = roc_auc_score(test_labels, test_probs[:, 1])

results_text = f"""
üéØ FINAL TEST RESULTS:
‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê
- Accuracy: {accuracy:.4f}
- ROC AUC Score: {roc_auc:.4f}

{report}
"""
print(results_text)
log_progress(results_text)

# Visualization
plt.figure(figsize=(20, 5))
plt.subplot(1, 4, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Val Loss')
plt.title('Loss vs. Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 4, 2)
plt.plot(history['train_acc'], label='Train Accuracy')
plt.plot(history['val_acc'], label='Val Accuracy')
plt.title('Accuracy vs. Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()

plt.subplot(1, 4, 3)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Normal', 'Pneumonia'], yticklabels=['Normal', 'Pneumonia'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.subplot(1, 4, 4)
fpr, tpr, _ = roc_curve(test_labels, test_probs[:, 1])
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.3f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()

plt.tight_layout()
plt.show()

log_progress("üéâ ALL PROCESSES COMPLETED SUCCESSFULLY!")
```
